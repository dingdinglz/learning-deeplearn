```python
from matplotlib import pyplot as plt
import torch
```

```python
def f(x):
    return x**2 + 1
def f2(x):
    return x-x + 1
```

```python
x = torch.arange(-5,5,0.1)
plt.plot(x,f(x),'-',label = 'f(x)')
plt.plot(x,f2(x),'-.',label = "f'(x)")
plt.plot(0,1,'.',markersize=15,label = 'meet')
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('wow')
plt.show()
```
