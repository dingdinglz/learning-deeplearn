```python
from matplotlib import pyplot as plt
import torch
```

```python
x = [1,2,3,4,5]
y = [1,8,27,64,125]
plt.plot(x,y,linewidth=5)
plt.xlabel('x')
plt.ylabel('x^3')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.title('折线绘制图')
plt.show()
```

```python
def f(x):
    return 3*x**2-4*x
```

```python
x = torch.arange(0,3,0.001)
plt.plot(x,f(x),'-',label='f(x)') # 画f(x)曲线
plt.plot(x,2*x-3,'r-.',label='Tangent line (x=1)') # 画f(x)的导数的线
plt.plot(1,-1,'g.',markersize=15,label='f(x) meet Tangent line') # 绘制一个点
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend() # 放置图例
plt.grid() # 放置方格
plt.title("f'(x)")
plt.savefig('test.png')
plt.show()
```
